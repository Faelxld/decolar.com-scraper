```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("Dados.csv")
df.head()
```

```python
df.shape
```

# Gráfico de distribuiçao dos preços das passagens
este grafico apresenta a distribuição dos preços das passagens no formato de ponto, onde se há a distinção dos dias da semana na qual foi realizada a pesquisa e com separação da parte do dia tambem(se foi realizada de Manha, tarde,noite ou madrugada)

```python
fig , axes = plt.subplots(figsize= (18,10))
plt.yticks(np.arange(0,3000,125))
sns.set_style('darkgrid')
sns.stripplot(x=df['week_day_search'],y=df['price'],hue = df['part_of_day'],palette='Set2')
axes.set_title("Distribuição do Preço das Passagens")
axes.set_ylabel("Preço das Passagens")
axes.set_xlabel('Dia da semana em que a Pesquisa foi realizada')
axes.legend(prop={'size': 12})
```

# Gráfico BoxPlot dos preços das passagens
este grafico apresenta a distribuição dos preços das passagens no formato Boxplot, onde se há a distinção dos dias da semana na qual foi realizada a pesquisa e com separação da parte do dia tambem(se foi realizada de Manha, tarde,noite ou madrugada)

```python
fig , axes = plt.subplots(figsize= (18,10))
plt.yticks(np.arange(0,3000,125))
sns.boxplot(x=df['week_day_search'],y=df['price'],hue = df['part_of_day'],palette='Set2')
axes.set_title("Distribuição BoxPlot das Passagens")
axes.set_ylabel("Preço das Passagens")
axes.set_xlabel('Dia da semana em que a Pesquisa foi realizada')
axes.legend(prop={'size': 12})
```

# Gráfico BarPlot dos preços das passagens
este grafico apresenta a a média dos preços das passagens no formato Boxplot, onde se há a distinção dos dias da semana na qual foi realizada a pesquisa e com separação da parte do dia tambem(se foi realizada de Manha, tarde,noite ou madrugada)

```python
fig , axes = plt.subplots(figsize= (18,10))
plt.yticks(np.arange(0,550,50))
sns.barplot(x=df['week_day_search'],y=df['price'],hue = df['part_of_day'],palette='Set2')
axes.set_title("Distribuição BoxPlot das Passagens")
axes.set_ylabel("Preço das Passagens")
axes.set_xlabel('Dia da semana em que a Pesquisa foi realizada')
axes.legend(prop={'size': 12})
```

# Gráfico ViolinPlot dos preços das passagens
este grafico apresenta a a média dos preços das passagens no formato Boxplot, onde se há a distinção dos dias da semana na qual foi realizada a pesquisa e com separação da parte do dia tambem(se foi realizada de Manha, tarde,noite ou madrugada)

```python
fig , axes = plt.subplots(figsize= (18,10))
plt.yticks(np.arange(0,3000,125))
sns.violinplot(x='week_day_search',y='price',data = df,hue = df['part_of_day'],palette='Set2')
axes.set_title("Distribuição BoxPlot das Passagens")
axes.set_ylabel("Preço das Passagens")
axes.set_xlabel('Dia da semana em que a Pesquisa foi realizada')
axes.legend(prop={'size': 12})
```
